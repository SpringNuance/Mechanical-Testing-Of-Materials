```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file into a DataFrame
data = pd.read_csv('results.csv', skiprows=[0])  # Assume header is in the first row now

# Plotting Force vs. Displacement
plt.figure(figsize=(10, 6))
plt.plot( data['ΔL [mm]'], data['Sigma_X [mm]'], label='Force vs. Displacement')
plt.ylabel('Force (mm)')
plt.title('Force vs. Displacement')
plt.legend()
plt.grid(True)
plt.show()

# Engineering Stress-Strain Curve
area_initial = 10 * 1  # initial cross-sectional area is 10 mm^2
engineering_stress = data['Sigma_X [mm]'] / area_initial  # Force/Area

plt.figure(figsize=(10, 6))
plt.plot(data['ΔL/L0 [1]'], engineering_stress, label='Engineering Stress-Strain Curve')
plt.xlabel('Strain (ΔL/L0)')
plt.ylabel('Stress (MPa)')
plt.title('Engineering Stress vs. Strain')
plt.legend()
plt.grid(True)
plt.show()

# True Stress-Strain Curve
# True stress = engineering stress * (1 + engineering strain)
# True strain = ln(1 + engineering strain)
true_stress = engineering_stress * (1 + data['ΔL/L0 [1]'])
true_strain = np.log(1 + data['ΔL/L0 [1]'])

plt.figure(figsize=(10, 6))
plt.plot(true_strain, true_stress, label='True Stress-Strain Curve')
plt.xlabel('True Strain')
plt.ylabel('True Stress (MPa)')
plt.title('True Stress vs. Strain')
plt.legend()
plt.grid(True)
plt.show()
```
