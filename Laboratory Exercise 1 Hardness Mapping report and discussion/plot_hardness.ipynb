```python
import matplotlib.pyplot as plt

# Data
point_numbers = list(range(1, 20))
hardness_values = [
    176.79, 200.12, 203.27, 195.34, 204.89, 203.81, 202.76,
    217.89, 242.14, 249.34, 253.09, 193.52, 194.52, 198.57,
    295.53, 295.51, 288.18, 294.57, 279.38
]

fontsize = 15
plt.figure(figsize=(8, 4))
plt.plot(point_numbers, hardness_values, marker='o', linestyle='-', color='blue')
plt.title('Hardness vs Point Number for Heat Affected Zone 1 (HAZ1)', fontsize=fontsize)
plt.xlabel('Point Number', fontsize=fontsize)
plt.ylabel('Hardness (HV1)', fontsize=fontsize)
plt.xticks(point_numbers, fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Adjusting the data for specific labels and color coding
labels = ['1st material (MA1)'] * 5 + ['1st heat affected zone (HAZ1)'] * 3 + ['Welding zone'] * 3 + ['2nd heat affected zone (HAZ2)'] * 3 + ['2nd material (MA2)'] * 5
colors = ['red'] * 5 + ['green'] * 3 + ['blue'] * 3 + ['purple'] * 3 + ['orange'] * 5

hardness_values = [
    176.79, 200.12, 203.27, 195.34, 204.89, 203.81, 202.76,
    217.89, 242.14, 249.34, 253.09, 193.52, 194.52, 198.57,
    295.53, 295.51, 288.18, 294.57, 279.38
]


# Plot
plt.figure(figsize=(8, 5))
for i, (x, y, label, color) in enumerate(zip(point_numbers, hardness_values, labels, colors)):
    plt.plot(x, y, marker='o', markersize=10, color=color, label=label if labels.index(label) == i else "")
plt.plot(point_numbers, hardness_values, color='black', linewidth=1)  # Line connecting the points

# Enhancing the plot
plt.title('Hardness value vs point number with different zones', fontsize=15)
plt.xlabel('Point Number', fontsize=15)
plt.ylabel('Hardness (HV1)', fontsize=15)
plt.xticks(point_numbers, fontsize=13)
plt.yticks(fontsize=13)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(fontsize=11)
plt.tight_layout()

plt.show()
```
